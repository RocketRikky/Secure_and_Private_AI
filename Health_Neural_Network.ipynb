```
import numpy as np
```

```
num_teachers = 10 #working with 10 hospitals
num_examples = 10000 
num_labels = 10
```

```
preds = (np.random.rand(num_teachers, num_examples) * num_labels).astype(int)
```

```
preds
```

```
preds[0].shape
```

```
preds[:,0]
# this is telling that the first element is predicting label 3, second element is predicting label 0 and so on
```

```
an_image = preds[:,0]
an_image
```

```
label_counts = np.bincount(an_image, minlength=num_labels)
# in an_image, 0 appeared 1 time and 2 appeared 2 times and so on.
```

```
epsilon = 0.1
beta = 1 / epsilon

for i in range(len(label_counts)):
  label_counts[i] += np.random.laplace(0, beta, 1)
```

```
label_counts
```

```
new_label = np.argmax(label_counts)
```

```
preds = (np.random.rand(num_teachers, num_examples) * num_labels).astype(int).transpose(1,0)
```

```
new_labels = list()
for an_image in preds:
  
  label_counts = np.bincount(an_image, minlength=num_labels)
  
  epsilon = 0.1
  beta = 1 / epsilon
  
  for i in range(len(label_counts)):
    label_counts[i] += np.random.laplace(0, beta, 1)
    
  new_label = np.argmax(label_counts)
  
  new_labels.append(new_label)
```

```
from syft.frameworks.torch.differential_privacy import pate
```

```
num_teachers, num_examples, num_labels = (100, 100, 10)
preds = (np.random.rand(num_teachers, num_examples) * num_labels).astype(int) #fake preds
true_indices = (np.random.rand(num_examples) * num_labels).astype(int) #true answers
```

```
data_dep_eps, data_inde_eps = pate.perform_analysis(teacher_preds=preds, indices=true_indices, noise_eps=0.1, delta=1e-5)
print("Data Independent Epsilon: ", data_inde_eps)
print("Data dependent Epsilon: ", data_dep_eps)
```

```
preds[:,0:5] *= 0
```

```
data_dep_eps, data_inde_eps = pate.perform_analysis(teacher_preds=preds, indices=true_indices, noise_eps=0.1, delta=1e-5)
print("Data Independent Epsilon: ", data_inde_eps)
print("Data dependent Epsilon: ", data_dep_eps)
```

```
preds[:,0:50] *= 0
```
